```python
image = io.imread("./bird.jpg")
image = rgb2gray(image)
image = random_noise(image, mode="s&p", amount=0.05)
show_images([image])
```

```python
'''
(1) Your own median filter algorithm
'''

def medianFilter(image, window_height=3, window_width=3):
    height, width = image.shape
    edge_x = math.floor(window_width / 2) 
    edge_y = math.floor(window_height / 2)

    outputImage = np.copy(image)

    for x in range(edge_x, width - edge_x):
        for y in range(edge_y, height - edge_y):
            colorArray = np.zeros((window_height, window_width))
            for fx in range(window_width):
                for fy in range(window_height):
                    colorArray[fy][fx] = image[y + fy - edge_y][x + fx - edge_x]
            colorArray.sort()
            outputImage[y][x] = colorArray[window_height // 2][window_width // 2]
    
    return outputImage

medianFilteredImage = medianFilter(image)
show_images([image, medianFilteredImage])
```

```python
'''
(2) skimage median filter
'''

medianFilteredImage2 = median(image)
show_images([image, medianFilteredImage2])
```

```python
'''
(3) Gaussion Filters
'''

gaussianFilteredImage1 = gaussian(image, sigma=8)
gaussianFilteredImage2 = gaussian(image, sigma=3)
gaussianFilteredImage3 = gaussian(image, sigma=0.2)

show_images([image, gaussianFilteredImage1, gaussianFilteredImage2, gaussianFilteredImage3])
```
